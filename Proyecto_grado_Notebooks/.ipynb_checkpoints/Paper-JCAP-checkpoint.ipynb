# Constants and unities: time (s), dist(Km)

```python
#Everything to Km and sec
Km = 1
seg= 1
pc = 3.0857e13*Km
Mpc= 1e6*pc 

#The values for Plank come from: https://arxiv.org/pdf/1807.06209.pdf (2018)
#The values for Riess come from: https://arxiv.org/pdf/1604.01424.pdf 
h  = 0.674                           #h of Hubble (Plank) H0=100*h
H0 = 100*h*Km/(seg*Mpc)              #H0 Plank
H0_err = 0.5*Km/(seg*Mpc)            #H0 error for Planck

H0_Union = 70*Km/(seg*Mpc)           #H0 Union2.1 
H0_Riess = 74.03*Km/(seg*Mpc)        #H0 Riess
H0_errR = 1.42*Km/(seg*Mpc)

Ok = 0                               #Omega k, assuming a flat universe
Om = 0.315                           #Omega matter
Ol = 1 - Om - Ok                     #Omega lambda
f  = Om**0.55                        #growth factor

C  = 299792458e-3*Km/seg             #speed of the light Km/seg
ly = 9.461e12*Km                     #light year
conv = np.pi/180                     #degree to radians 

print('The age of the Universe is',H0**(-1)/(3600*24*365)*1e-6, 'Millions of years')
```

# Distances in cosmology 

```python
#Distancia comovil
def Dist_com(z, O_matter=Om, O_k=0, H_Hubble=H0):
    #H(z) de las ecuaciones de Friedmann
    #Hubble = H0*( Om*(1.0 + z)**3.0 + Ok*(1.0 + z)**2.0 + Ol )**0.5
    
    #dist comovil
    #Ol = 1-Om-Ok
    if type(z)==float: z = np.array([z])
    dist_com = np.array([C*sp.integrate.quad(lambda z:1/( (O_matter*(1 + z)**3 + O_k*(1 + z)**2 + (1 - O_matter - O_k) )**0.5 ), 0, zi)[0]/H_Hubble for zi in z])
    #dist_com = C*sp.integrate.quad(lambda z:1.0/Hubble,0.0,z)[0]

    return dist_com


#Diameter angular distance (or physical distance) 
def Dist_ang(z, O_matter=Om, O_k=0, H_Hubble=H0):
    #Factor de escala
    z = np.array(z)
    a = 1/(1 + z)
    
    #Distancia de luminosidad
    dist_ang = Dist_com(z, O_matter, O_k, H_Hubble)*a
    
    return dist_ang


#Distancia de luminosidad D_l con el redshift
def Dist_lz(z, O_matter=Om, O_k=0, H_Hubble=H0):
    #Factor de escala
    z = np.array(z)
    a = 1/(1 + z)
    
    #Distancia de luminosidad
    dist_lum = Dist_com(z, O_matter, O_k, H_Hubble)/a

    return dist_lum


#Distancia usando el modulo de la distancia, se convierte de pc a km.
#Esta distancia corresponde a la distancia de luminosidad D_l
#mu_shift es la corrección debida a que Union2.1 utiliza H=H_Union para el distance modulus
def Dist_lm(mu, shift=False, H_Hubble=H0, H_2=H0_Union):
    if shift:
        mu_shift = 5*np.log10(H_2/H_Hubble)
        mu_corr  = np.array(mu) - mu_shift
        dist_lum = 10**(mu_corr/5 + 1)
    
    else: dist_lum = 10**(np.array(mu)/5 + 1)
    
    return dist_lum*pc


#H(z):
def HubbleE(z, O_matter=Om, O_k=0, H_Hubble=H0):
    z = np.array(z)
    Hubble_ = H_Hubble*( O_matter*(1 + z)**3 + O_k*(1 + z)**2 + (1 - O_matter - O_k) )**0.5
    return Hubble_
```

# Radial velocity(peculiar) and the Hubble flow 

```python
#Calculo de la componente de la velocidad visible a nosotros
#Hubble Trouble or Hubble bubble
#Dl es el valor teorico calculado con dist mod, Dl_ el calculado con z.
def vel(Dl, z, O_matter=Om, O_k=0, H_Hubble=H0):
    Dl_ = Dist_lz(z, O_matter, O_k, H_Hubble)
    return -C*np.array(z)*( np.array(Dl)/np.array(Dl_) - 1)


#The Hubble flow
def Vel_Hubble_flow(z, O_matter=Om, O_k=0, H_Hubble=H0):
    #H_z = H0*( Om*(1.0 + z)**3.0 + Ok*(1.0 + z)**2.0 + Ol )**0.5
    return H_Hubble*Dist_ang(z, O_matter, O_k, H_Hubble)
```

# Statistics

```python
#Error in the distance computed using the distance modulus.
def Err_dist_mod(Dis_, Err_):
    error_ = np.log(10)*np.array(Dis_)*np.array(Err_)/5               #Valor de error en la distancia
    error_porc = 100*np.array(error_)/np.array(Dis_)                  #Valor porcentual del error
    return np.array([error_, error_porc])


#Weighted average
#This formula is derived from the Likelihood method for a data where each x_i follows a
#gaussian destribution with a different error (standar deviation) for each x_i
def Weighted_mean(data, Errs):
    data = np.array(data)
    Errs = np.array(Errs)
    
    weighted_mean = np.sum(data/Errs**2.0)/np.sum(Errs**(-2))
    return weighted_mean


#Weighted standar deviation
#This formula is derived from the Likelihood method for a data where each x_i follows a
#gaussian destribution with a different error (standar deviation) for each x_i
#I guess this formula is good for large values of n (number of elements to average)
def Weighted_std(Errs):
    Errs = np.array(Errs)
    
    weighted_var = 1/np.sum(1.0/Errs**2)
    return np.sqrt(weighted_var)


#Error in the radial velocitie (model)
#H_err is the error in the Hubble parameter
#H_Hubble is the Hubble parameter
#modD is the distance modulus and modDE is its error
def Error_Vr(modD, modDE, shift=False, H_Hubble=H0, H_2=H0_Union, H_err=H0_err):
    modD = np.array(modD)
    modDE = np.array(modDE)
    
    #Error in the distance computed using de distance modulus.
    Err_dist_mu = Err_dist_mod(Dis_= Dist_lm(mu=modD, shift=shift, H_Hubble=H_Hubble, H_2=H_2), Err_=modDE)[0]
    Dl_mu = Dist_lm(mu=modD, shift=shift, H_Hubble=H_Hubble, H_2=H_2)
    
    #This formula for the error is calculated by using the aproximation dl=cz/H0
    return np.sqrt( (H_Hubble*Err_dist_mu)**2 + (Dl_mu*H_err)**2 )


#Dnh is the luminosity distance from GW  for a non-homogenious universe
#eDnh is the error in the luminosity distance for a non-homoginious universe
def Error_GWVr(Dnh,eDnh, H_Hubble=H0, H_err=H0_err):
    #This formula for the error is calculated by using the aproximation dl=cz/H0
    return np.sqrt( (H_Hubble*eDnh)**2 + (Dnh*H_err)**2 )
    
    
    
#Chi² for a lineal model ax+b.
#data_x and data_y are the data in the x and y axis respectively.
#a0 and b0 are initial guess for the parameters a and b.
#limits are the ranges in which a and b will be found, if no list is passed 
#the limits will be assumed to be (-1,1) for each parameter. The list has
#the structure limits=[(inf_a, sup_a),(inf_b,sup_b)].
#error_x and error_y are the errors in each of the axis x and y. If there's not errors
#then the minimization will be done without weights. 
def chi2_lineal_minuit(data_x, data_y, a0, b0, error_x=False, error_y=False, limits=False, print_level=0, minos=False):
    ##################################################################
    ##################################################################
    #Assigning errors as weights for the chi2_func function
    #this set of "if's" conditionals is to make sure that there's
    #always a weight for the chi2_func function    
    if type(error_x)==bool or type(error_y)==bool:
        #If only one is boolean there's error only for x or y:  
        if type(error_x)==bool:
            errX = np.zeros_like(data_x)
        else:
            errX = error_x
            
        if type(error_y)==bool:
            errY = np.zeros_like(data_y)
        else:
            errY = error_y
    #If no one is boolean
    else:
        errX = error_x
        errY = error_y
        
    #If both are boolean theres not errors
    if type(error_x)==bool and type(error_y)==bool:
        errX = np.ones_like(data_x)
        errY = np.zeros_like(data_y)
        
    #Defining the chi2 function for a lineal model   
    def chi2_func(a,b):
        model_ = a*data_x + b
        chi2 = np.sum( (data_y - model_)**2/np.sqrt(errX**2 + errY**2)**2 )
        return chi2 
    ##################################################################
    ##################################################################
    
    #The limits for the parameters a and b
    if limits:
        inf_a = limits[0][0]
        sup_a = limits[0][1]
        inf_b = limits[1][0]
        sup_b = limits[1][1]
    else:
        inf_a = -1
        sup_a = 1
        inf_b = -1
        sup_b = 1
    
    #Minimizing chi2_func using minuit
    result_minuit = Minuit(chi2_func, a=a0, error_a=0.1, limit_a=(inf_a,sup_a),
                                      b=b0, error_b=0.1, limit_b=(inf_b,sup_b),
                                      errordef=1, print_level = print_level)
    result_minuit_mingrad = result_minuit.migrad()
    
    if minos:
        result_minuit.minos();
        result_minuit.print_param()
        
    #The values of the parameters and their errors
    a = result_minuit.values[0]
    b = result_minuit.values[1]
    aE = result_minuit.errors[0]
    bE = result_minuit.errors[1]
    
    #The chi2 minimal value
    chi2_min = chi2_func(a,b)
    print("The minimize X² is =", chi2_min)
    
    return [(a,aE),(b,bE)]
```

# Shells

```python
##############################################################################################################
##############################################################################################################
#This function calculate the shells for a list "data"
#If limits=True then it must be passed a list "bounds" wiht the limits.
#If n is an integer then it will build up n shells, if it is a float
#it will be the thickness of the shell in the same unities as "data".
def Shells_simple(Data, n, limits=False, bounds=None):
    Data = np.array(Data)
    
    #Determining if it is any interval to build the shells
    if limits:
        d_min = bounds[0]
        d_max = bounds[1]
    else:
        d_min = 0
        d_max = max(Data)  
    
    #The step
    if type(n)==int: 
        step = (d_max - d_min)/n
        #print('int',step)
    else:
        step = n
        #print('float',step)
        
    #List to save the boolean list for each shell and the distance per shell
    shell_idx = list()
    shell_dist = list()
    
    #Building the shells
    d_lim = d_max
    dist=0.0
    while dist < d_lim:
        dist = d_min + step
        idx_shell = (d_min<Data) & (dist>Data)
        if len(Data[idx_shell]) !=0:
            shell_idx.append(idx_shell)
            shell_dist.append(Data[idx_shell])
        d_min = dist
    #print('Shells=',len(shell_idx))
    return [shell_idx, shell_dist, len(shell_idx)]
```

# Redshift correction with 2M++

```python
def zCorrection(tmppCoords, tmppV, SN_z, SN_RAC, SN_DEC, zCorr=False, method='mean', O_m=Om, O_k=Ok, H_Hubble=H0, crit=True):
    #If you don't wanna do the redshift correction just set zCorr=False
    if zCorr==False: return [np.ones_like(SN_z, dtype=bool) , np.zeros_like(SN_z)]
    
    #If you want to
    else:
        h = H_Hubble/100*(seg*Mpc/Km)
        if crit: crit=(400/256)*Mpc/h
        else: crit=crit
    
        #For 2M++ the coordinates are cartesian galactic comovil
        #Now we just take those galaxies within a radio of 200/h Mpc
        R_tmpp = np.linalg.norm(tmppCoords/h, axis=1)
        idxR_tmpp200h = 200.0/h > R_tmpp
        tmppCoords = tmppCoords[idxR_tmpp200h]*Mpc/h
        tmppV = tmppV[idxR_tmpp200h]

        #SNs
        d_com = Dist_com(z=SN_z, H_Hubble=H_Hubble, O_matter=O_m, O_k=O_k)
        coors_icrs_SN = SkyCoord(ra=SN_RAC*u.deg, dec=SN_DEC*u.deg, distance=d_com*u.km, frame='icrs')
        coors_gala_SN = coors_icrs_SN.transform_to('galactic')

        #Cartesian coordinates for the SN in galactic coordinates
        XYZ_SN = np.array(coors_gala_SN.cartesian.get_xyz()).T


        #Comparing SN and tmpp
        SN_idx = []
        Vtmpp  = []
        Ctmpp  = []
        for i in range(len(XYZ_SN)):
            idx_crit = (abs(XYZ_SN[i][0]-tmppCoords.T[0])<crit) & (abs(XYZ_SN[i][1]-tmppCoords.T[1])<crit) & (abs(XYZ_SN[i][2]-tmppCoords.T[2])<crit) 
            if len(tmppCoords[idx_crit])==0: 
                SN_idx.append(False)
                Vtmpp.append(False)
                Ctmpp.append(False)
            else: 
                SN_idx.append(True)
                Vtmpp.append(tmppV[idx_crit])
                Ctmpp.append(tmppCoords[idx_crit])

        SN_idx = np.array(SN_idx)
        Vtmpp  = np.array(Vtmpp)
        Ctmpp  = np.array(Ctmpp)

        #The redshift correction is V_tmpp/c
        if method=='mean':
            return [SN_idx, np.array([np.average(Vtmpp_i) for Vtmpp_i in Vtmpp[SN_idx]])/C ]
            #return [SN_idx, XYZ_SN, np.array(Ctmpp), np.array([np.average(Vtmpp_i) for Vtmpp_i in Vtmpp])/C ]

        if method=='closest':
            zCclosest = []
            for i in range(len(XYZ_SN[SN_idx])):
                if SN_idx[SN_idx][i]:
                    closestId = Closest_point(list_coor=Ctmpp[SN_idx][i], point=XYZ_SN[SN_idx][i])
                    zCclosest.append( (Vtmpp[SN_idx][i])[closestId]/C )
            return [SN_idx, np.array(zCclosest)]
            #return [SN_idx, XYZ_SN, np.array(Ctmpp), Vtmpp ,np.array(zCclosest)]

        if method=='All':  
            return [SN_idx, XYZ_SN, Ctmpp, Vtmpp/C]
```

# Another functions

```python
################################################arctang function##################################################
#it returns the arctan(y/x) taken into account the quadrant in which the point is. It returns the angle in radians
#for the spesial point (0,0) it returns angle=0 radians
def arctanInt(x,y):
    #Idx for the coordinates
    idx = np.array(range(len(x)))
    
    #Computing the normal arctan
    arctan = np.arctan(y/x)
    
    #Choosing the quadrant
    quadrant1 = (0<x) & (0<y)
    quadrant2 = (0>x) & (0<y)
    quadrant3 = (0>x) & (0>y)
    quadrant4 = (0<x) & (0>y)
    
    #Idx for those points in each quadrant
    idxq1 = idx[quadrant1]
    idxq2 = idx[quadrant2]
    idxq3 = idx[quadrant3]
    idxq4 = idx[quadrant4]
    
    #Spesial points
    X0_Yn = (0==np.array(x)) & (0>y)                #x=0 and y<0 -> angle=270 degrees
    Xn_Y0 = (0>x) & (0==np.array(y))                #x<0 and y=0 -> angle=180 degrees
    X0_Y0 = (0==np.array(x)) & (0==np.array(y))     #x=0 and y=0 -> there is not angle (we define 0 degrees)
    
    #Spesial points
    id_X0_Yn = idx[X0_Yn]               
    id_Xn_Y0 = idx[Xn_Y0] 
    id_X0_Y0 = idx[X0_Y0]  
    
    #Changing the arctan for each quadrant 
    #For the first quadrant it is not necesssary make any change
    for i in idxq2:
        arctan[i] = arctan[i] + np.pi
    for j in idxq3:
        arctan[j] = arctan[j] + np.pi
    for k in idxq4:
        arctan[k] = arctan[k] + 2*np.pi 
    
    #Special points
    for l in id_X0_Yn:
        arctan[l] = 3*np.pi/2
    for m in id_Xn_Y0:
        arctan[m] = np.pi
    for n in id_X0_Y0:
        arctan[n] = 0
        
    return arctan


#This function returns a list without the Nan values in the list list_
def NoNanInf(list_):
    no_nan = [i for i in list_ if ~(np.isnan(i) or np.isinf(i)) ]
    return np.array(no_nan)


#This function returns the closest point (3D->[x_i,y_i,z_i]) to "point" in the list "list_coor"
def Closest_point(point,list_coor, idx=True):
    if idx: 
        return distance.cdist([point],np.array(list_coor)).argmin()
    else:
        return list_coor[ distance.cdist([point],np.array(list_coor)).argmin() ]


#This function returns the closest value to "value" in the list "lists_"
def Closest_value(value, list_):
    return (list_[np.absolute(value-np.array(list_)).argmin()], np.absolute(value-np.array(list_)).argmin())


#A lineal mode for the adjust
def Lineal_Model(x,a,b):
    return a*x + b
```

```python
#The maximal redshift we are working with. This value is based in the depth of 2M++ which is 0.067.
max_zPR = 0.0685

#SNs with z<max_z
idxshellPR = z_pantheonR<max_zPR

#SNs from pantheon for the shells
z_corrPR = [np.ones_like(z_pantheonR[idxshellPR], dtype=bool), np.zeros_like(z_pantheonR[idxshellPR])]
Zs_PR_shell = z_pantheonR[idxshellPR][z_corrPR[0]] + z_corrPR[1]

#Building the shells for pantheon
#n_shellP = 29
n_shellPR = 0.004
Shells_PR = Shells_simple(Data=Zs_PR_shell, n=n_shellPR)
Shells_PRidx = Shells_PR[0]

#Information for the shells in Pantheon
z_shellPR   = np.array([Zs_PR_shell[idxP] for idxP in Shells_PRidx])          #z's for each shell
za_shellPR  = np.array([np.mean(i) for i in z_shellPR])                       #z average for each shell
zE_shellPR  = np.array([np.std(i) for i in z_shellPR])                        #std for z in each shell
len(za_shellPR), len(Zs_PR_shell) 
```

```python
#Calculating the average density contrast with pantheon (Riess)from a X² 

#With shells
def chi_2_shellPR(a, b, O_m, H):
    #Lineal model
    model = a*za_shellPR + b
    
    #Error in the distance computed with the distance modulus
    E_DisMod_shell = Err_dist_mod(Dis_=Dist_lm(mu=mu_pantheonR, shift=False), Err_=muE_pantheonR)[0][idxshellPR][z_corrPR[0]]
    
    #Dl->luminosity distance with mu, Dl_->luminosity distance with z
    Dlmu_shell = Dist_lm(mu=mu_pantheonR, shift=False)[idxshellPR][z_corrPR[0]] #Dl with mu for SN in pantheon whith z<z_max
    Dl = np.array([Weighted_mean(data=Dlmu_shell[shell_i], Errs=E_DisMod_shell[shell_i]) for shell_i in Shells_PRidx])
    DlE= np.array([Weighted_std(Errs=E_DisMod_shell[shell_i]) for shell_i in Shells_PRidx])
    
    #The radial velocity for each shell
    V_shell = vel(Dl=Dl, z=za_shellPR, O_matter=O_m, H_Hubble=H)
    VE_shell= np.sqrt((H*DlE)**2 + (Dl*H0_err)**2)
    
    chi2 = np.sum( ((V_shell/C-model)/np.sqrt(zE_shellPR**2 + (VE_shell/C)**2))**2 )
    return chi2

#Without shells
def chi_2_nshellPR(a, b, O_m, H):
    #Lineal model
    model = a*Zs_PR_shell + b
    Dlmu = Dist_lm(mu=mu_pantheonR, shift=False)[idxshellPR][z_corrPR[0]]
    
    #Radial velocity for each SN with z<z_max
    V_nshell  = vel(Dl=Dlmu, z=Zs_PR_shell, O_matter=O_m, H_Hubble=H) 
    E_vr_nshell = Error_Vr(modD=mu_pantheonR, modDE=muE_pantheonR, shift=False, H_Hubble=H)[idxshellPR][z_corrPR[0]]
    
    chisq2 = np.sum( ( (V_nshell/C - model)/(E_vr_nshell/C)  )**2.0 )
    return chisq2
```

```python
#X² for a piecewise function using shells for Pantheon (Riess) 
def chi2_z011_shell_zlimPR(a, z_lim, O_m, H):
    #Zones 
    Idx_zona1 = z_lim>z_0z11PR
    Idx_zona2 = z_lim<z_0z11PR

    #Lineal model
    model1 = a*z_0z11PR[Idx_zona1] 
    
    #Velocities for SNs and and the GW's
    GWv  = vel(Dl=GW_dl, z=GW_z, H_Hubble=H, O_matter=O_m)    
    vP = vel(Dl=Dist_lm(mu=mu_pantheonR, shift=False), z=z_pantheonR, O_matter=O_m, H_Hubble=H)
    #vP = np.concatenate( (vP, GWv) )
    
    #Errors for the velocities
    GWvE = Error_GWVr(Dnh=GW_dl, eDnh=GW_dlE, H_Hubble=H)
    vPE  = Error_Vr(modD=mu_pantheonR, modDE=muE_pantheonR, shift=False, H_Hubble=H)
    #vPE  = np.concatenate( (vPE, GWvE) )
    
    #Velocities and errors for each shell
    v_0z11 = np.array([Weighted_mean(data=vP[j], Errs=vPE[j]) for j in shellsId_0z11PR[0]])
    vE_0z11= np.array([Weighted_std(Errs=vPE[j]) for j in shellsId_0z11PR[0]])
    
    #X² zone 1
    chi2_zona1 = np.sum(  (v_0z11[Idx_zona1]/C - model1)**2/(vE_0z11[Idx_zona1]/C)**2 )

    #The second region is where the density contrast averaged begins to drop like 1/r³   
    model2 = a*z_lim**3/(z_0z11PR[Idx_zona2])**2
    chi2_zona2 = np.sum( (v_0z11[Idx_zona2]/C - model2)**2/(vE_0z11[Idx_zona2]/C)**2 )

    return chi2_zona1 + chi2_zona2


#X² for a piecewise function with Pantheon (Riess) without using shells
def chi2_z011_nshell_zlimPR(a, z_lim, O_m, H):
    #Zones
    Idx_zona1 = z_lim>z_0z11nPR
    Idx_zona2 = z_lim<z_0z11nPR
    
    #Lineal model
    model1 = a*z_0z11nPR[Idx_zona1] 
    
    #Velocities for SNs and and the GW's
    GWv  = vel(Dl=GW_dl, z=GW_z, H_Hubble=H, O_matter=O_m)    
    vP = vel(Dl=Dist_lm(mu=mu_pantheonR, shift=False), z=z_pantheonR, O_matter=O_m, H_Hubble=H)
    vP = vP[idx_nShellPR]
    #v = np.concatenate( (v, GWv) )[idx_nShellP]
    
    #Errors for the velocities
    GWvE = Error_GWVr(Dnh=GW_dl, eDnh=GW_dlE, H_Hubble=H)
    vPE = Error_Vr(modD=mu_pantheonR, modDE=muE_pantheonR, shift=False, H_Hubble=H)
    vPE = vPE[idx_nShellPR]
    #vPE = np.concatenate( (vPE, GWvE) )[idx_nShellP]
    
    chi2_zona1 = np.sum( (vP[Idx_zona1]/C - model1)**2/(vPE[Idx_zona1]/C)**2 )
    
    #The second region is where the density contrast averaged begins to drop like 1/r³   
    model2 = a*z_lim**3/(z_0z11nPR[Idx_zona2])**2
    chi2_zona2 = np.sum( (vP[Idx_zona2]/C - model2)**2/(vPE[Idx_zona2]/C)**2 )
     
    return chi2_zona1 + chi2_zona2
```

# Union 2.1 (Riess)

```python
#The 2M++ catalogue has a deep of 0.06 (z), so we choose only those SN with z<max_z for 
#building the shells.
#The maximal redshift we are working with. This value is based in the depth of 2M++ which is 0.067.
max_zUR = 0.0685

#SNs with z<max_z
idxshellUR = z_union<max_zUR

#The z correction
z_corrUR = zCorrection(tmppCoords=tmppCoords, tmppV=tmppV, 
                     SN_RAC=ACR[idxshellUR], SN_DEC=DEC[idxshellUR], SN_z=z_union[idxshellUR], 
                     method='mean', zCorr=False, 
                     H_Hubble=H0_Riess, O_m=Om, crit=True)

#SNs to apply the z correction
Zs_SN_shellUR = z_union[idxshellUR][z_corrUR[0]] + z_corrUR[1]  

#Building the shells for Union 2.1
n_shellUR = 29
Shells_SNUR = Shells_simple(Data=Zs_SN_shellUR, n=n_shellUR)

Shells_SNidxUR = Shells_SNUR[0]

#Information for the shells in Union 2.1
z_shellUR   = np.array([Zs_SN_shellUR[idxSN] for idxSN in Shells_SNidxUR])    #z's for each shell
za_shellUR  = np.array([np.mean(i) for i in z_shellUR])                       #z average for each shell
zE_shellUR= np.array([np.std(i) for i in z_shellUR])                          #std for z in each shell
len(za_shellUR), len(Zs_SN_shellUR)
```

```python
#Calculating the average density contrast from a X²

#With shells
def chi_2_shell_snUR(a, b, O_m, H):
    #Lineal model
    model = a*za_shellUR + b
    
    #Error in the distance computed with the distance modulus
    E_DisMod_shell = Err_dist_mod(Dis_=Dist_lm(mu=dist_mod_union, shift=True, H_Hubble=H), Err_=dist_mod_unionE)[0][idxshellUR][z_corrUR[0]]
    
    #Dl->luminosity distance with mu, Dl_->luminosity distance with z
    Dlmu_shell = Dist_lm(mu=dist_mod_union, shift=True, H_Hubble=H)[idxshellUR][z_corrUR[0]] #Dl with mu for SN thith z<z_max
    Dl = np.array([Weighted_mean(data=Dlmu_shell[shell_i], Errs=E_DisMod_shell[shell_i]) for shell_i in Shells_SNidxUR])
    DlE= np.array([Weighted_std(Errs=E_DisMod_shell[shell_i]) for shell_i in Shells_SNidxUR])
    
    #The radial velocity for each shell
    V_shell = vel(Dl=Dl, z=za_shellUR, O_matter=O_m, H_Hubble=H)
    VE_shell= np.sqrt((H*DlE)**2 + (Dl*H0_err)**2)
    
    chi2 = np.sum( ((V_shell/C-model)/np.sqrt(zE_shellUR**2 + (VE_shell/C)**2))**2 )
    return chi2

#Without shells
def chi_2_nshell_snUR(a, b, O_m, H):
    #Lineal model
    model = a*Zs_SN_shellUR + b
    Dlmu = Dist_lm(mu=dist_mod_union, shift=True, H_Hubble=H)[idxshellUR][z_corrUR[0]]
    
    #Radial velocity for each SN with z<z_max
    V_nshell  = vel(Dl=Dlmu, z=Zs_SN_shellUR, O_matter=O_m, H_Hubble=H) 
    E_vr_shell = Error_Vr(modD=dist_mod_union, modDE=dist_mod_unionE, shift=True, H_Hubble=H)[idxshellUR][z_corrUR[0]]
    
    chisq2 = np.sum( ( (V_nshell/C - model)/(E_vr_shell/C)  )**2.0 )
    return chisq2
```

```python
#Minimization using Minuit
#For SNe with shells
result_shell_minuitUR = Minuit(chi_2_shell_snUR, a=0,   error_a=0.1,  limit_a=(-1,1),
                                                 b=0,   error_b=0.1,  limit_b=(-1,1),
                                                 O_m=Om, error_O_m=0.1, limit_O_m=(0,1), fix_O_m=True,
                                                 H=H0_Riess,   error_H=H0_Riess/10, limit_H=(0,1e-17), fix_H=True,
                                                 errordef=1)

#For SNe without shells
result_nshell_minuitUR = Minuit(chi_2_nshell_snUR, a=0,   error_a=0.1,  limit_a=(-1,1),
                                                   b=0,   error_b=0.1,  limit_b=(-1,1),
                                                   O_m=Om, error_O_m=0.1, limit_O_m=(0,1), fix_O_m=True,
                                                   H=H0_Riess, error_H=H0_Riess/10, limit_H=(0,1e-17), fix_H=True,
                                                   errordef=1)
```

# Beyond $z_{max}$, determining the size of the inhomogeneity with Union 2.1  (Riess) $0<z<z_{sup}$

```python
ns_iUR  = 0.004      #This is the step in "unities" of z to build the shells
z_infUR = 0.0       #lower z
#z_supUR = 0.3

#For shells
shellsId_0z11UR = Shells_simple(Data=z_union_full, n=ns_iUR, limits=True, bounds=[z_infUR, z_supUR])
z_0z11UR = np.array([np.mean(z_union_full[j]) for j in shellsId_0z11UR[0]])
Sn_0z11_namesUR = np.array([names_union_full[j] for j in shellsId_0z11UR[0]])

#Without shells
idx_0z11_fullUR  = z_union_full<z_supUR
z_0z11_fullUR = z_union_full[idx_0z11_fullUR]
Sn_0z11_names_fullUR = names_union_full[idx_0z11_fullUR]
len(z_0z11UR), len(z_0z11_fullUR)
```

```python
#X² for a piecewise function using shells 
def chi2_z011_shell_zlimUR(a, z_lim, O_m, H):
    #Zones 
    Idx_zona1 = z_lim>z_0z11UR
    Idx_zona2 = z_lim<z_0z11UR
    
    #Lineal model
    model1 = a*z_0z11UR[Idx_zona1] 
    
    #Velocities for SNs and and the GW's
    GWv  = vel(Dl=GW_dl, z=GW_z, H_Hubble=H, O_matter=O_m)    
    v_full = vel(Dl=Dist_lm(mu=mod_union_full, shift=True, H_Hubble=H), z=z_union_full[0:-len(GW_z)], O_matter=O_m, H_Hubble=H)
    v_full = np.concatenate( (v_full, GWv) )
    
    #Errors for the velocities
    vr_unionE_full = Error_Vr(modD=mod_union_full, modDE=dist_mod_unionE_full, shift=True, H_Hubble=H)
    GWvE = Error_GWVr(Dnh=GW_dl, eDnh=GW_dlE, H_Hubble=H)
    vr_unionE_full = np.concatenate( (vr_unionE_full, GWvE) )
    
    #Velocities and errors for each shell
    v_0z11 = np.array([Weighted_mean(data=v_full[j], Errs=vr_unionE_full[j]) for j in shellsId_0z11UR[0]])
    vE_0z11= np.array([Weighted_std(Errs=vr_unionE_full[j]) for j in shellsId_0z11UR[0]])
    
    chi2_zona1 = np.sum(  (v_0z11[Idx_zona1]/C - model1)**2/(vE_0z11[Idx_zona1]/C)**2 )
        
    #The second region is where the density contrast averaged begins to drop like 1/r³   
    model2 = a*z_lim**3/(z_0z11UR[Idx_zona2])**2
    chi2_zona2 = np.sum( (v_0z11[Idx_zona2]/C - model2)**2/(vE_0z11[Idx_zona2]/C)**2 )
     
    return chi2_zona1 + chi2_zona2

#X² for a piecewise function without using shells
def chi2_z011_nshell_zlimUR(a, z_lim, O_m, H):
    #Zones
    Idx_zona1 = z_lim>z_0z11_fullUR
    Idx_zona2 = z_lim<z_0z11_fullUR
    
    #Lineal model
    model1 = a*z_0z11_fullUR[Idx_zona1] 
    
    #Velocities for SNs and and the GW's
    GWv  = vel(Dl=GW_dl, z=GW_z, H_Hubble=H, O_matter=O_m)    
    v_full = vel(Dl=Dist_lm(mu=mod_union_full, shift=True, H_Hubble=H), z=z_union_full[0:-len(GW_z)], O_matter=O_m, H_Hubble=H)
    v_full = np.concatenate( (v_full, GWv) )[idx_0z11_fullUR]
    
    #Errors for the velocities
    vr_unionE_full = Error_Vr(modD=mod_union_full, modDE=dist_mod_unionE_full, shift=True, H_Hubble=H)
    GWvE = Error_GWVr(Dnh=GW_dl, eDnh=GW_dlE, H_Hubble=H)
    vr_unionE_full = np.concatenate( (vr_unionE_full, GWvE) )[idx_0z11_fullUR]
    
    chi2_zona1 = np.sum( (v_full[Idx_zona1]/C - model1)**2/(vr_unionE_full[Idx_zona1]/C)**2 )
        
    #The second region is where the density contrast averaged begins to drop like 1/r³   
    model2 = a*z_lim**3/(z_0z11_fullUR[Idx_zona2])**2
    chi2_zona2 = np.sum( (v_full[Idx_zona2]/C - model2)**2/(vr_unionE_full[Idx_zona2]/C)**2 )
     
    return chi2_zona1 + chi2_zona2                                      
```

```python
result_minuit_z011_shell_zlimUR = Minuit(chi2_z011_shell_zlimUR, a=0.02, error_a=0.001,  
                                                                 z_lim=0.07, error_z_lim=0.001,
                                                                 limit_z_lim=(0.0, z_supUR),
                                                                 O_m=Om, error_O_m=0,   fix_O_m=True,
                                                                 H=H0_Riess, error_H=0, fix_H=True,
                                                                 errordef=1.0)

result_minuit_z011_nshell_zlimUR = Minuit(chi2_z011_nshell_zlimUR, a=-0.053, error_a=0.001, 
                                                                   z_lim=0.0, error_z_lim=0.001,
                                                                   limit_z_lim=(0.0, z_supUR), 
                                                                   O_m=Om, error_O_m=0, fix_O_m=True,
                                                                   H=H0_Riess, error_H=0, fix_H=True,
                                                                   errordef=1.0)
```
