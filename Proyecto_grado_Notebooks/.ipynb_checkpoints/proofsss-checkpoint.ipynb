```python
def indexf(value,l0):
    #Almacenamiento
    Idx = []
    idx = 0

    #El ciclo rompe al acabar de examinar la lista
    while True:
        
        try:
            #index(value,idx) da el indice de "value" a partir de idx
            idx = l0.index(value,idx)
            Idx.append(idx)
            idx=idx+1
        except:break
            
    return Idx

def arctanInt(x,y):
    #Idx for the coordinates
    idx = np.array(range(len(x)))
    
    #Computing the normal arctan
    arctan = np.arctan(y/x)
    
    #Choosing the quadrant
    quadrant1 = (0.0<x) & (0.0<y)
    quadrant2 = (0.0>x) & (0.0<y)
    quadrant3 = (0.0>x) & (0.0>y)
    quadrant4 = (0.0<x) & (0.0>y)
    
    #Idx for those points in each quadrant
    idxq1 = idx[quadrant1]
    idxq2 = idx[quadrant2]
    idxq3 = idx[quadrant3]
    idxq4 = idx[quadrant4]
    
    #Spesial points
    X0_Yn = (0.0==np.array(x)) & (0.0>y)                #x=0 and y<0 -> angle=270 degrees
    Xn_Y0 = (0.0>x) & (0.0==np.array(y))                #x<0 and y=0 -> angle=180 degrees
    X0_Y0 = (0.0==np.array(x)) & (0.0==np.array(y))     #x=0 and y=0 -> there is not angle (we define 0 degrees)
    
    #Spesial points
    id_X0_Yn = idx[X0_Yn]               
    id_Xn_Y0 = idx[Xn_Y0] 
    id_X0_Y0 = idx[X0_Y0]  
    
    #Changing the arctan for each quadrant 
    #For the first quadrant it is not necesssary make any change
    for i in idxq2:
        arctan[i] = arctan[i] + np.pi
    for j in idxq3:
        arctan[j] = arctan[j] + np.pi
    for k in idxq4:
        arctan[k] = arctan[k] + 2.0*np.pi 
    
    #Spesial points
    for l in id_X0_Yn:
        arctan[l] = 3*np.pi/2.0
    for m in id_Xn_Y0:
        arctan[m] = np.pi
    for n in id_X0_Y0:
        arctan[n] = 0.0
        
    return arctan
```

```python
np.random.seed(1234)
n_lim = 30
n_dat = 100
xs = np.linspace(0,4,n_dat)+0.01*np.random.rand(n_dat)

xlim = xs[n_lim]
a = -1.1
b = a*xlim**4.0


ys1 = a*xs[0:n_lim]
ys2 = b*xs[n_lim:]**(-3.0)
ys = np.concatenate((ys1, ys2))+0.02*np.random.rand(len(xs))
ysE= 0.01*np.random.rand(len(xs))
len(xs), len(ys), xlim
```

```python
#Fit
def chi2_pieseWise(a_, xlim_):
    idx_z1 = xlim_>xs
    idx_z2 = xlim_<xs
    
    #z1 
    model1 = a_*xs[idx_z1]
    chi2_1 = np.sum( (model1 - ys[idx_z1])**2/(ysE[idx_z1])**2 )
    
    #z2
    model2 = a_*xlim_**4/xs[idx_z2]**3
    chi2_2 = np.sum( (model2 - ys[idx_z2])**2/(ysE[idx_z2])**2 )
    
    return chi2_1 + chi2_2
```

```python
def rot(W, i, w, f):
    xs = np.cos(f)*( np.cos(W)*np.cos(w)-np.sin(W)*np.sin(w)*np.cos(i) ) + np.sin(f)*( -np.cos(W)*np.sin(w)-np.sin(W)*np.cos(i)*np.cos(w) )
    ys = np.cos(f)*( np.sin(W)*np.cos(w)+np.cos(W)*np.cos(i)*np.sin(w) ) + np.sin(f)*( -np.sin(W)*np.sin(w)+np.cos(W)*np.cos(i)*np.cos(w) )
    zs = np.cos(f)*np.sin(i)*np.sin(w) + np.sin(f)*np.sin(i)*np.sin(w)
    return[xs, ys, zs]
```

```python
AU = 1.0
pi = np.pi
W = np.array([48.331, 76.68, 348.7393, 49.562, 100.492, 113.6428, 73.989, 131.7943 ])*pi/180
i = np.array([7.004, 3.394, 1.5786, 1.850, 1.3053, 2.4845, 0.77255, 1.7680 ])*pi/180
w = np.array([29.124, 55.156, 114.2078, 286.537, 275.066, 336.0138, 96.591, 265.6469 ])*pi/180
a = np.array([0.387, 0.7273, 1.0, 1.5237, 5.2042, 9.5820, 19.2294, 30.1037]) 
e = np.array([0.2056, 0.00677, 0.01671, 0.0933, 0.04839, 0.05648, 0.04441, 0.008585])
```

```python
def orbitas(W, i, w, a, e):
    b = a*np.sqrt(1 - e**2)     #semieje menor
    
    #Matriz de rotaciÃ³n usando Ã¡ngulos de Euler
    rot_w = np.array([ [np.cos(w), np.sin(w), 0], [-np.sin(w), np.cos(w), 0], [0, 0, 1] ])
    rot_i = np.array([ [1, 0, 0], [0, np.cos(i), np.sin(i)], [0, -np.sin(i), np.cos(i)] ])
    rot_W = np.array([ [np.cos(W), -np.sin(W), 0], [np.sin(W), np.cos(W), 0], [0, 0, 1] ])
    rot_total = (rot_w.dot(rot_i)).dot(rot_W)
    #rot_total = np.linalg.inv(rot_total)
    
    #Coordenadas cartesianas de la Ã³rbita
    fs = np.linspace(0, 2*pi, 100)
    xs = a*np.sin(fs)
    ys = b*np.cos(fs)
    zs = np.zeros_like(fs)
    rs = np.array([xs, ys, zs])
    
    #Ã“rbita rotada segÃºn los Ã¡ngulos de Euler
    Xs = np.zeros_like(fs)
    Ys = np.zeros_like(fs)
    Zs = np.zeros_like(fs)
    
    for j in range(len(fs)):
        Rrot = rot_total.dot( np.array([xs[j], ys[j], zs[j]]) )
        Xs[j] = Rrot[0]
        Ys[j] = Rrot[1]
        Zs[j] = Rrot[2]
        
    Rs = [Xs, Ys, Zs]
    
    ax.plot(Rs[0]/AU, Rs[1]/AU, Rs[2]/AU, '-')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    
    return 
```

```python
#GrÃ¡fica
fig = plt.figure(figsize=(7,7))
ax = Axes3D(fig)

for j in range(len(W)-4):
    orbitas(W[j], i[j], w[j], a[j], e[j])
```
